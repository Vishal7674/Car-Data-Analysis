Step5:- Data Cleaning

1.Find all null values in dataset and fill with mean of that column.(Data cleaning)

```python
# Find null values in given dataset
df.isnull().sum()
```

```python
#Check unique values in Make cloumn
df.EngineSize.unique()
```

```python
#Find mean of Make column
x =df.EngineSize.mean()
x
```

```python
# Fill na values in EngieSize column by its mean value 
df['EngineSize'].fillna(x, inplace =True)
```

```python
df['EngineSize'].isnull().sum()
```

```python
#Simillerly we can fill all Na values from all columns by its mean values 
df['Weight'].fillna(df['Weight'].mean(), inplace =True)
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace = True)
df['Horsepower'].fillna(df['Horsepower'].mean(), inplace = True)
df['MPG_City'].fillna(df['MPG_City'].mean(), inplace = True)
df['MPG_Highway'].fillna(df['MPG_Highway'].mean(), inplace = True)
df['Weight'].fillna(df['Weight'].mean(), inplace = True)
df['Wheelbase'].fillna(df['Wheelbase'].mean(), inplace = True)
df['Length'].fillna(df['Length'].mean(), inplace = True)
```

```python
#Check again for any null value in data set
df.isnull().sum()
```

```python
#Remaining null value cloumns are containing string datatypes so we use mode value to replace na values 
df['Make'].fillna(df['Make'].mode() , inplace =True)
df['Model'].fillna(df['Model'].mode(), inplace =True)
df['Type'].fillna(df['Type'].mode(), inplace = True)
df['Origin'].fillna(df['Origin'].mode(), inplace = True)
df['DriveTrain'].fillna(df['DriveTrain'].mode(), inplace = True)
df['Invoice'].fillna(df['Invoice'].mode(), inplace = True)
df['MSRP'].fillna(df['MSRP'].mode(), inplace = True)
```

```python
#Check again is there any null value presents
df.isnull().sum()
```
